# K-Means

K-Means Clustering in Machine Learning: Deep Dive
K-Means clustering is one of the most popular unsupervised machine learning algorithms. It is used for partitioning a dataset into K distinct, non-overlapping groups (also called clusters) based on the similarity of the data points. The goal of K-Means is to find natural groupings within the data and minimize the intra-cluster variance.

1. What is Clustering?
Clustering is a type of unsupervised learning where the algorithm groups data points based on similarity. Each group or cluster contains data points that are more similar to each other than to those in other clusters. K-Means clustering achieves this by optimizing the position of the cluster centroids.

2. Key Concepts of K-Means:
Centroid: A centroid is the center point of a cluster. Each cluster has a centroid that represents the average of all data points within the cluster.

K (Number of Clusters): K is the predefined number of clusters that the algorithm will partition the dataset into. K is a hyperparameter that you must set before running the algorithm, but finding the optimal K is one of the challenges in K-Means.

Euclidean Distance: K-Means typically uses the Euclidean distance metric to measure the distance between data points and centroids. The distance is calculated between each data point and the centroid of each cluster.

3. Working of K-Means Algorithm:
The K-Means algorithm works iteratively through the following steps:

Initialization:

The first step is to initialize K centroids randomly (or using other initialization techniques like K-Means++). These centroids represent the initial centers of each of the K clusters.
Assignment Step:

Each data point is assigned to the nearest centroid based on the chosen distance metric (typically Euclidean distance). Every point is assigned to the cluster whose centroid is closest to it.
Update Step:

After assigning all the points to clusters, the centroids are updated. The new centroid of each cluster is the mean (average) of all the points assigned to that cluster. This step recalculates the center of each cluster.
Repeat:

Steps 2 and 3 are repeated iteratively. The assignment step and update step are carried out until the centroids no longer change significantly or the algorithm reaches a predefined maximum number of iterations. When convergence occurs (i.e., the centroids stabilize), the algorithm stops.

5. Choosing the Optimal K (Number of Clusters):
One of the major challenges in K-Means is determining the optimal number of clusters (K). A few common methods for selecting K include:

Elbow Method:
The idea is to run the K-Means algorithm for a range of K values and plot the within-cluster sum of squares (inertia) against K. The "elbow" point in the plot represents the optimal number of clusters where the rate of decrease in inertia slows down.
Silhouette Score:
This method evaluates how similar each point is to its own cluster compared to other clusters. A higher silhouette score indicates well-separated and well-defined clusters. You can experiment with different K values and choose the one that maximizes the silhouette score.

Simplicity and Speed: K-Means is easy to implement and is computationally efficient, especially for large datasets. It has a time complexity of

Interpretability: The final clusters are easy to understand since each data point is assigned to a cluster, and each cluster is represented by a centroid.

7. Disadvantages of K-Means:
Sensitivity to Initialization: The initial random selection of centroids can lead to suboptimal solutions. To mitigate this, algorithms like K-Means++ have been introduced, which select better initial centroids.

K Needs to Be Predefined: You must specify the number of clusters (K) in advance, which can be challenging if there is no prior knowledge of the dataset.

Assumes Spherical Clusters: K-Means assumes that clusters are spherical and equally sized. It struggles with clusters of different shapes and densities or when the clusters are not well-separated.

Sensitive to Outliers: K-Means is sensitive to outliers because they can heavily influence the position of centroids.

8. Variants of K-Means:
K-Means++: A smarter initialization method that improves the initialization of centroids by spreading them out before starting the algorithm. This reduces the likelihood of poor local minima and leads to better clustering results.

Mini-Batch K-Means: Instead of using the entire dataset for each iteration, Mini-Batch K-Means uses a small subset (batch) of the data. This makes the algorithm more scalable and faster, especially for large datasets.

K-Medoids: A variation of K-Means where the centroids are actual data points (medoids) instead of the mean of the data points. This method is more robust to outliers.

9. Applications of K-Means:
Customer Segmentation: In marketing, K-Means is often used to segment customers based on purchasing behavior, demographics, or other attributes.

Image Compression: K-Means can be used to reduce the number of colors in an image by clustering similar colors together.

Anomaly Detection: K-Means can help detect anomalies in data by identifying points that do not fit well into any cluster.

Document Clustering: It can group similar documents together based on content, enabling better search and recommendation systems.

10. Practical Considerations:
Preprocessing: K-Means requires the data to be normalized or standardized, especially when the features are on different scales (e.g., age and income).

Cluster Size: The algorithm assumes that all clusters have approximately the same size, which can be problematic if clusters have very different sizes or densities.

Convergence Issues: K-Means can converge to local minima and may not always find the global optimum solution, especially if the initialization is poor.

11. Conclusion:
K-Means clustering is a widely used and powerful algorithm for unsupervised learning tasks, such as data segmentation. Despite its simplicity, it has some limitations, including the need to predefine the number of clusters and its sensitivity to initial centroids and outliers. By understanding these limitations and leveraging advanced techniques like K-Means++ and Mini-Batch K-Means, you can mitigate some of these challenges and make the algorithm work more effectively for your specific use case.

# WCSS
In the context of K-Means clustering, WCSS stands for Within-Cluster Sum of Squares, which is a metric used to evaluate the compactness of the clusters formed during the K-Means algorithm. It measures the variance within each cluster, representing how close the data points in each cluster are to the cluster's centroid. A smaller WCSS indicates that the data points are more tightly grouped around the centroids, suggesting more compact and well-separated clusters.

2. Interpretation of WCSS:
Lower WCSS: A lower WCSS value indicates that the data points are closer to their respective centroids, which generally suggests well-formed and compact clusters.
Higher WCSS: A higher WCSS value indicates that the data points are spread out over a larger area, meaning the clusters are not as tight, and the model may not be capturing the underlying structure of the data effectively.
3. WCSS and the Elbow Method:
One common use of WCSS is in determining the optimal number of clusters (K) when applying the K-Means algorithm. This is done using the Elbow Method.

The Elbow Method involves running K-Means clustering for a range of

The point where the rate of decrease starts to slow down significantly (forming an "elbow") represents the optimal number of clusters. This is the point at which adding more clusters no longer significantly improves the compactness of the clusters.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/mall.csv")
df.head()
```

```python
df = df.drop(columns =['CustomerID', 'Genre'])
df.head()
```

```python
x = df.iloc[:,[0,1]].values
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
a = []
for i in range(1,11):
  b = KMeans(n_clusters = i, init = 'k-means++', random_state =42)
  b.fit(x)
  a.append(b.inertia_)

plt.plot(range(1,11),a)
plt.title('The Elbow method graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

# from the above plot, we can see the elbow point is at 4. So the number of clusters here will be 4.

```python
b = KMeans(n_clusters = 4, init= 'k-means++',random_state =42)
y_predict = b.fit_predict(x)
```

```python
# Visualaizing the clusters
plt.scatter(x[y_predict==0,0],x[y_predict==0,1], s =100,c = 'blue', label = 'cluster 1') # for the cluster 1
plt.scatter(x[y_predict == 1,0], x[y_predict == 1,1], s =100, c = 'green', label = 'cluster 2') # for the cluster 2
plt.scatter(x[y_predict == 2,0],x[y_predict== 2,1], s =100, c ='red',label = 'cluster 3') # for the cluster 3
plt.scatter(x[y_predict==3,0],x[y_predict ==3,1], s = 100, c = 'cyan', label = 'cluster 4') # for the cluster 4

plt.scatter(b.cluster_centers_[:,0],b.cluster_centers_[:,1], s =300, c = 'yellow', label = 'Centroid')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
