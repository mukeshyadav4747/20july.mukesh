#What is Principal Component Analysis (PCA)?

>>> by chatgpt

Principal Component Analysis (PCA) is a statistical technique used to simplify a dataset by transforming it into a new set of orthogonal (uncorrelated) variables called principal components. These principal components capture the maximum variance in the data while reducing the dimensionality. PCA is a linear transformation that finds the directions (or axes) in which the data varies the most.

Key Points of PCA:
Dimensionality Reduction: PCA is commonly used for reducing the number of features (dimensions) in a dataset, making it easier to analyze and visualize the data, without losing significant information.
Variance Maximization: It finds the directions (principal components) along which the data varies the most. The first principal component captures the most variance, the second captures the second most, and so on.
Data Simplification: By projecting the data onto the first few principal components, we can retain most of the important information while discarding noise or less significant features.
Uncorrelated Features: PCA transforms correlated features into uncorrelated ones. The new features (principal components) are linear combinations of the original features.

Advantages of PCA:
Dimensionality Reduction: PCA reduces the number of features (dimensions), making it easier to visualize and analyze the data.
Noise Reduction: By removing less significant components (those with low variance), PCA helps in removing noise and irrelevant features.
Improves Performance: For machine learning algorithms, reducing dimensionality can improve performance, speed up training, and reduce the risk of overfitting.
Uncorrelated Features: The principal components are uncorrelated, which simplifies analysis and models.
Disadvantages of PCA:
Linear Assumption: PCA assumes a linear relationship between the features. It might not perform well on data that has nonlinear structures.
Interpretability: The principal components are combinations of the original features, which can make interpretation difficult.
Sensitive to Scaling: Features with large scales or different units can dominate the principal components if the data is not standardized.
Applications of PCA:
Data Visualization: PCA is often used to reduce high-dimensional data to 2D or 3D for visualization.
Noise Filtering: It can help remove noise by discarding components with low variance.
Feature Extraction: PCA can be used to extract the most important features from a dataset.
Preprocessing for Machine Learning: PCA is commonly used as a preprocessing step to reduce dimensionality before applying machine learning algorithms like clustering, classification, and regression.
In summary, PCA is a powerful tool for simplifying high-dimensional datasets while preserving their most important features. It is widely used in various fields like data science, image processing, and pattern recognition.





```python
import numpy as np
import pandas as pd
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
df.head()
```

```python
df = df.drop(columns=['User ID','Gender'])
df
```

```python
x = df.drop(columns=['Purchased'], axis =1)
y =df['Purchased']
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

pipe = Pipeline([
    ('scaler', StandardScaler()),('pca', PCA(n_components=2)),('classifier',RandomForestClassifier(n_estimators=100
                                                                                                   ,random_state=42))
])
```

```python
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
acc = accuracy_score(y_test, y_pred)
print(acc)
```
