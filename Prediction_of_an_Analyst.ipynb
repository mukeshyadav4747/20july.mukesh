```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = np.asanyarray(df[['YearsExperience']])
y = np.asanyarray(df[['Salary']])
```

```python
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state =42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
a = float(input('Years of Experience'))
features = np.array([[a]])
print('Predicted Salary=', lr.predict(features))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
a = float(input('Years of Experience'))
features = np.array([[a]])
print('Predicted Salary=', dt.predict(features))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train,y_train)
```

```python
a = float(input('Years of Experience'))
features = np.array([[a]])
print('Predicted Salary=', rf.predict(features))
```
