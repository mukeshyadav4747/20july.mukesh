# RandomForest Model
A Random Forest is a powerful ensemble learning algorithm used in machine learning for both classification and regression tasks. It operates by building a collection of decision trees during training time and outputs the class that is the mode (in classification) or the mean (in regression) of the individual trees' predictions.

Key Concepts:
Ensemble Method: Random Forest is an ensemble technique, which means it combines the predictions of multiple individual models (in this case, decision trees) to make more accurate predictions than any individual model could.

Decision Trees: The base models in Random Forest are decision trees, which are simple models that split the data based on feature values to make predictions.

```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('/content/Social_Network_Ads.csv')
df.head(3)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['Gender'] = lb.fit_transform(df['Gender'])
df.head(3)
```

```python
x = df.drop(columns=['Purchased'])
y= df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
