```python
 import numpy as np
 import pandas as pd
```

```python
df = pd.read_csv("/content/covid_toy.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer   # to fill the data into missing data
```

```python
si = SimpleImputer(strategy="mean")
```

```python
df['fever'] = si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head(3)
```

```python
df2 =df.copy()
df2.head(3)
```

```python
df2['gender']= df2['gender'].map({'Male':1,'Female':0})
```

```python
df2.sample(3)
```

```python
df2['cough']= df2['cough'].map({'Mild':0,'Strong':1})
df2['city']= df2['city'].map({'Kolkata':0,'Banglore':1,'Delhi':2,'Mumbai':3})
df2['has_covid']=df2['has_covid'].map({'Yes':1,'No':0})
```

```python
df['city'].value_counts()
```

```python
df2.head(3)
```

```python
x= df2.drop(columns=['has_covid'])   ## input data
y= df2['has_covid']  ## target data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.2,random_state = 42)
```

```python
np.round(x_train.describe(),1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)  ## fit means learn the parameter and transform means apply it on data.
```

```python
x_train_new = pd.DataFrame(x_train_sc,columns = x_train.columns)
```

```python
np.round(x_train_new.describe(),1)   # here 1 denotes to 0 after the point
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn =mn.fit_transform(x_train)
```

```python
x_train_new =pd.DataFrame(x_train_mn, columns= x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```
